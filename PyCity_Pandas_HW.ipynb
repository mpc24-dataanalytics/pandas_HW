```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

```

## Top Performing Schools (By Passing Rate)

```python
school_data_complete.head()
```

```python
school_type=school_data.set_index('school_name')['type']
total_school_students=school_data_complete['school_name'].value_counts()
total_school_budget=school_data_complete.groupby('school_name').mean()['budget']
per_student_budget=total_school_budget/total_school_students
avg_per_school_math=school_data_complete.groupby('school_name').mean()['math_score']
avg_per_school_reading=school_data_complete.groupby('school_name').mean()['reading_score']
school_passing_math=school_data_complete[school_data_complete['math_score']>=70]
school_passing_reading=school_data_complete[school_data_complete['reading_score']>=70]
per_school_passing_math=school_passing_math.groupby('school_name').count()['student_name']/total_school_students*100
per_school_passing_reading=school_passing_reading.groupby('school_name').count()['student_name']/total_school_students*100
overall_passing=(per_school_passing_math + per_school_passing_reading)/2
```

```python
school_summary=pd.DataFrame({
                            'School Type':school_type,
                            'Total Students':total_school_students,
                            'Total School Budget':total_school_budget,
                            'Per Student Budget':per_student_budget,
                            'Average Math Score':avg_per_school_math,
                            'Average Reading Score':avg_per_school_reading,
                            '% Passing Math':per_school_passing_math,
                            '% Passing Reading':per_school_passing_reading,
                            'Overall Passing Rate':overall_passing})
school_summary
```

```python
top_5 = school_summary.sort_values('Overall Passing Rate',ascending=False)
top_5.head()            
```
